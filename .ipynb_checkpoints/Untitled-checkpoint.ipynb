## Power on

```python
import matplotlib.pyplot as plt
from datetime import datetime

with open("./logs/rittal_PowerOn.log", "r") as f:
    # initialize empty lists to store the values and timestamps
    values1_r = []
    values2_r = []
    values3_r = []
    values4_r = []
    timestamps_r = []
    
    for line in f:
        # split the line into parts and convert the values to float
        parts = line.strip().split()
        voltageRit = float(parts[4])
        currentRit = float(parts[5])
        powerRit = float(parts[6])
        energyRit = float(parts[7])
        
        # extract the timestamp from the line and convert it to a datetime object
        timestamp_str = parts[0] + ' ' + parts[1]
        timestamp_r = datetime.strptime(timestamp_str, '%Y-%m-%d %H:%M:%S.%f')
        
        # append the values and timestamp to their respective lists
        values1_r.append(voltageRit)
        values2_r.append(currentRit)
        values3_r.append(powerRit)
        values4_r.append(energyRit)
        timestamps_r.append(timestamp_r)
    
#     plot each value in a separate graph
    fig1, ax1 = plt.subplots()
    ax1.plot(timestamps_r, values1_r, 'bo', markersize=1)
    ax1.set_xlabel('Time')
    ax1.set_ylabel('Voltage')
    ax1.set_ylim(bottom=0)
    fig2, ax2 = plt.subplots()
    ax2.plot(timestamps_r, values2_r, 'go', markersize=1)
    ax2.set_xlabel('Time')
    ax2.set_ylabel('Current')
    ax2.set_ylim(bottom=0)
    fig3, ax3 = plt.subplots()
    ax3.plot(timestamps_r, values3_r, 'ro', markersize=1)
    ax3.set_xlabel('Time')
    ax3.set_ylabel('Power')
    ax3.set_ylim(bottom=0)
    fig4, ax4 = plt.subplots()
    ax4.plot(timestamps_r, values4_r, 'yo', markersize=1)
    ax4.set_xlabel('Time')
    ax4.set_ylabel('Energy')
    ax4.set_ylim(bottom=0)
    plt.show()

```

## Wattsup 

```python
import matplotlib.pyplot as plt
from datetime import datetime

# read the values and timestamps from the wattsup_PowerOn.log file
with open("./logs/wattsup_PowerOn.log", "r") as f:
    # initialize empty lists to store the values and timestamps
    values1_ws = []
    values2_ws = []
    values3_ws = []
    timestamps_ws = []
    
    for line in f:
        # split the line into parts and convert the values to float
        parts = line.strip().split()
        powerWat = float(parts[3])
        voltageWat = float(parts[4])
        currentWat = float(parts[5])
        
        # extract the timestamp from the line and convert it to a datetime object
        timestamp_str = parts[0] + ' ' + parts[1]
        timestamp = datetime.strptime(timestamp_str, '%Y-%m-%d %H:%M:%S.%f')
        
        # append the values and timestamp to their respective lists
        values1_ws.append(powerWat)
        values2_ws.append(voltageWat)
        values3_ws.append(currentWat)
        timestamps_ws.append(timestamp)

# plot each value in a separate graph
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
axs[0].plot(timestamps_ws, values2_ws, 'ro', markersize=1)
axs[0].set_xlabel('Time')
axs[0].set_ylabel('Voltage')
axs[0].set_ylim(bottom=0)
axs[0].set_title('Voltage')

axs[1].plot(timestamps_ws, values3_ws, 'go', markersize=1)
axs[1].set_xlabel('Time')
axs[1].set_ylabel('Current')
axs[1].set_ylim(bottom=0)
axs[1].set_title('Current')

axs[2].plot(timestamps_ws, values1_ws, 'bo', markersize=1)
axs[2].set_xlabel('Time')
axs[2].set_ylabel('Power')
axs[2].set_ylim(bottom=0)
axs[2].set_title('Power')

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(15,8))

# plot the voltage, current, and power values from the second code in the top row
axs[0,0].plot(timestamps_ws, values2_ws, 'ro', markersize=1)
axs[0,0].set_xlabel('Time')
axs[0,0].set_ylabel('Voltage')
axs[0,0].set_ylim(bottom=0)
axs[0,0].set_title('Voltage_Watsup')

axs[0,1].plot(timestamps_ws, values3_ws, 'go', markersize=1)
axs[0,1].set_xlabel('Time')
axs[0,1].set_ylabel('Current')
axs[0,1].set_ylim(bottom=0)
axs[0,1].set_title('Current_Watsup')

axs[0,2].plot(timestamps_ws, values1_ws, 'bo', markersize=1)
axs[0,2].set_xlabel('Time')
axs[0,2].set_ylabel('Power')
axs[0,2].set_ylim(bottom=0)
axs[0,2].set_title('Power_Watsup')

# plot the voltage, current, and power values from the first code in the bottom row
axs[1,0].plot(timestamps_r, values1_r, 'bo', markersize=1)
axs[1,0].set_xlabel('Time')
axs[1,0].set_ylabel('Voltage')
axs[1,0].set_ylim(bottom=0)
axs[1,0].set_title('Voltage')

axs[1,1].plot(timestamps_r, values2_r, 'go', markersize=1)
axs[1,1].set_xlabel('Time')
axs[1,1].set_ylabel('Current')
axs[1,1].set_ylim(bottom=0)
axs[1,1].set_title('Current')

axs[1,2].plot(timestamps_r, values3_r, 'ro', markersize=1)
axs[1,2].set_xlabel('Time')
axs[1,2].set_ylabel('Power')
axs[1,2].set_ylim(bottom=0)
axs[1,2].set_title('Power')

plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,5))

# plot the voltage values in one graph
ax1.plot(timestamps_r, values2_r, 'bo', markersize=1, label='Voltage - Rittal')
ax1.plot(timestamps_ws, values2_ws, 'ro', markersize=1, label='Voltage - Wattsup')
ax1.set_xlabel('Time')
ax1.set_ylabel('Voltage')
ax1.legend()

# plot the current values in one graph
ax2.plot(timestamps_r, values3_r, 'bo', markersize=1, label='Current - Rittal')
ax2.plot(timestamps_ws, values3_ws, 'ro', markersize=1, label='Current - Wattsup')
ax2.set_xlabel('Time')
ax2.set_ylabel('Current')
ax2.legend()

# plot the energy values in one graph
ax3.plot(timestamps_r, values1_r, 'bo', markersize=1, label='Energy - Rittal')
ax3.plot(timestamps_ws, values1_ws, 'ro', markersize=1, label='Energy - Wattsup')
ax3.set_xlabel('Time')
ax3.set_ylabel('Energy')
ax3.legend()

plt.show()
```
